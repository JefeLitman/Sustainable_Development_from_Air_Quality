```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import bokeh
import os
```

# Carga de datos Anuales en el archivo 2019_05

```python
path = "/content/gdrive/My Drive/Bogota/medicionesAire"
path = "data/medicionesAnuales/TodosDatosAnuales.xlsx"

TotalData = pd.read_excel(path)
TotalData["indicador"] = TotalData["Nombre del indicador"].astype(str).map(str) + " (" +TotalData["Unidades del indicador"].astype(str) + ")"
TotalData.set_index("indicador",inplace=True)
```

```python
#Los datos tuvo que realizarse unos pequeños cambios en el excel porque algunos datos estaban coo fechas (Index 73 y 2)
#Y cambiar el nombre de la dimension a una forma sin tildes (económico)
TotalData.head()
```

```python
#Extraccion de los dataframe de las dimensiones
for i, g in TotalData.groupby(by='Dimensión'):
    globals()[str(i)] =  g
```

# Dimension economica

```python
económico.fillna(value=0,inplace=True)
económico.drop("Dimensión",axis=1,inplace=True)
económico.drop("Nombre del indicador",axis=1,inplace=True)
económico.drop("Unidades del indicador",axis=1,inplace=True)
económico.drop("No,",axis=1,inplace=True)
```

```python
económico.head()
```

### Graficas de los indicadores

```python
plt.figure(figsize=(30,20))
for i in range(económico.shape[0]):
  plt.subplot(6,3,i+1)
  económico.iloc[i].plot(title=económico.index.values[i])
```

### Correlacion entre los indicadores

```python
corr = económico.transpose().corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
